```python
# Code to import libraries
import pandas as pd
dataframe=pd.read_excel("sample_data/Female and Male Astronauts Accepted in the NASA Program.xlsx")
```

```python
# Delection of column ID
dataframe=dataframe.drop(dataframe.columns[[0]],axis=1)
dataframe
```

```python
# Printing column names
list(dataframe.columns.values)
```

```python
# Checking for NA in dataset
dataframe.isna()
```

```python
# Getting the count of missing data
dataframe.isna().sum()
```

```python
# Counting Males and Females in the column count
dataframe["count"]=dataframe['Males Accepted in NASA Program ']+dataframe['Females Accepted in NASA Program ']
```

```python
# Printing the data
dataframe
```

```python
# Creating a new dataframe before the year 2017
newdataframe=dataframe[dataframe["Year"]<=2017]
```

```python
# Creating a new dataframe after the year 1990
newdataframe=newdataframe[newdataframe["Year"]>=1990]
newdataframe
```

```python
# Histogram plot for Females Accepted in NASA Program
import matplotlib.pyplot as plt
plt.hist(newdataframe['Females Accepted in NASA Program '])
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("Females Accepted in NASA Program histogram")
```

```python
# Histogram plot for Males Accepted in NASA Program
plt.hist(newdataframe['Males Accepted in NASA Program '])
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("Males Accepted in NASA Program histogram")
```

```python
#Bar plot for the the Males and Females Accepted in the NASA Program 
plt.bar(newdataframe['Year'],newdataframe['Males Accepted in NASA Program '],label="Males")
plt.bar(newdataframe['Year'],newdataframe['Females Accepted in NASA Program '],label="Females")
plt.xlabel("Year")
plt.ylabel("count")
plt.title("Males and Females Accepted in NASA Program histogram")
plt.legend()
```

The data is taken from year 1959 to 2022, it gives the details of males and females accepted in the NASA program. Until 1969 there was no females admitted in the NASA program; after 2017 there was no admission to the NASA program which admits astronauts within a 2 to 3 year period but due to onset of COVID there are no record admissions into the NASA program. The overall admission into NASA is 100 every year.Males were expected more until 2010 but after 2010 Females were accepted more frequently.
