```python
#Linear Regression of Males by Year
%matplotlib inline
import  statsmodels.api as sm
import statsmodels.stats.api as sms
from scipy.stats import boxcox
x = dataframe['Males Accepted in NASA Program ']
y = dataframe['Year']
model = sm.OLS(y,x).fit()
model.summary()
```

The R-square values shows that males count is dependent on year. Higher the R-square valur the higher the correlation.

```python
# Linear Regression of Females by Year
%matplotlib inline
import  statsmodels.api as sm
import statsmodels.stats.api as sms
from scipy.stats import boxcox
x = dataframe['Females Accepted in NASA Program ']
y = dataframe['Year']
model = sm.OLS(y,x).fit()
model.summary()
```

The R-square shows that the year and females are not correlated; as the value is 0.4 which is far from 1, which is a weak correlation .
