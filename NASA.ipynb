## Remove extra columns to be able to combine into one data frame 

```python
dataframe1=dataframe1.drop(columns=['ID','DOB','Education','rando','Announcement Date','Age','insamp'])
```

```python
## convert 
def recode (Space_Flights):
    if Space_Flights >= 1:
        return 1
    if Space_Flights == 0:
        return 0
dataframe2['Space Flights R'] = dataframe2['Space Flights'].apply(recode)
dataframe2.head()
```

```python
print(dataframe3.info())
```

```python
dataframe2=dataframe2.drop(columns=['Group','Year','Space Flights','Space Flight (hr)','Space Walks','Space Walks (hr)','Space Flights R'])
```

```python
NasaData=pd.merge(dataframe1,dataframe2, on='Gender')
```

```python
NasaData.head()
```

```python
def recode (Gender):
    if Gender == "Male":
        return 1
    if Gender == "Female":
        return 0
NasaData['GenderR'] = NasaData['Gender'].apply(recode)
```
