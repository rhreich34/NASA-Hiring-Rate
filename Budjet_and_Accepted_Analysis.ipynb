```python
#1.	What is the difference between men getting accepted versus women?
dataframe.sample(n=30,random_state=42)
```

```python
# Single Sample t-test for Females
import scipy.stats as stats
#perform one sample t-test
stats.ttest_1samp(a=dataframe['Females Accepted in NASA Program '], popmean=30)
```

1. The null hypothesis , mean = 30
2. Alternaye hypothesis , the mean is not equal to 30 results. The p-value is less than 0.05 and the null hypothesis is rejected and the alternate hypothesis is correct.


```python
# Single Sample t-test for Females
import scipy.stats as stats
#perform one sample t-test
stats.ttest_1samp(a=dataframe['Males Accepted in NASA Program '], popmean=50)
```

1. The null hypothesis , mean = 50
2. Alternaye hypothesis , the mean is not equal to 50 results. The p-value is less than 0.05 and the null hypothesis is rejected and the alternate hypothesis is correct.

```python
#Linear Regression of Females versus Budget
%matplotlib inline
import  statsmodels.api as sm
import statsmodels.stats.api as sms
from scipy.stats import boxcox
x = dataframe['Females Accepted in NASA Program ']
y = dataframe['White House Budget Submission']
model = sm.OLS(x,y).fit()
model.summary()
```
